**Loading the Data File**

The data for the publication statistics extracted from: https://publons.com/researcher/1923625/ahmet-altun/metrics/

```python
df = pd.read_excel("./Altun-Pub-Stat.xlsx")
df
```

**Plot**

```python
plt.style.use('ggplot') 
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(df["Year"], df["Publications"], color='cornflowerblue')
ax2.plot(df["Year"], df["Times_Cited"], color='mediumblue', linewidth=2)

ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Publications', color='cornflowerblue')
ax2.set_ylabel('Times Cited', color='mediumblue')
ax1.set_xticklabels(df["Year"], rotation=45)
plt.xticks(df["Year"])
ax1.set_xlim(left=2000.2, right=2021.8)
ax1.set_ylim(bottom=0, top=10)
ax2.set_ylim(bottom=0, top=250)
plt.grid();
```
